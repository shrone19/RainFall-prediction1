```python
df = df.drop(columns=['maxtemp', 'temparature', 'mintemp'])
```

```python
df.head()
```

```python
df['rainfall'].value_counts()
```

```python
df_major = df[df['rainfall']==1]
df_minor = df[df['rainfall']==0]
```

```python
print(df_major.shape)
print(df_minor.shape)
```

```python
from sklearn.utils import resample
```

```python
df_major_under =  resample(df_major,replace=False,n_samples=len(df_minor),random_state=45)
```

```python
df_major_under.shape
```

```python
df_undersampled = pd.concat([df_major_under,df_minor])
```

```python
df_undersampled.shape
```

```python
df_undersampled.head()
```

```python

df_undersampled = df_undersampled.sample(frac=1, random_state=45).reset_index(drop=True)
```

```python
df_undersampled.head()
```

```python
df_undersampled['rainfall'].value_counts()
```

```python
x = df_undersampled.drop(columns=['rainfall'])
y = df_undersampled['rainfall']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=45)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
ypred_r = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,ypred_r))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_features": ["sqrt", "log2"],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}
```

```python
grid_search_rf = GridSearchCV(estimator=rfc, param_grid=param_grid_rf, cv=5, n_jobs=-1, verbose=2)
```

```python
grid_search_rf.fit(x_train,y_train)
```

```python
best_rfc = grid_search_rf.estimator
```

```python
best_rfc_pred = best_rfc.predict(x_test)
```

```python
print(classification_report(y_test,best_rfc_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
ypred_log = lr.predict(x_test) 
```

```python
print(classification_report(y_test,ypred_log))
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, ComplementNB , MultinomialNB
```

```python
gnb = GaussianNB()
bnb = BernoulliNB()
cnb = ComplementNB()
mnb = MultinomialNB()
```

```python
gnb.fit(x_train,y_train)
bnb.fit(x_train,y_train)
cnb.fit(x_train,y_train)
mnb.fit(x_train,y_train)
```

```python
gnb_pred = gnb.predict(x_test)
bnb_pred = bnb.predict(x_test)
cnb_pred = cnb.predict(x_test)
mnb_pred = mnb.predict(x_test)
```

```python
print('GaussianNB')
print(classification_report(y_test,gnb_pred))

print('BernoulliNB')
print(classification_report(y_test,bnb_pred))

print('ComplementNB')
print(classification_report(y_test,cnb_pred))

print('MultinomialNB')
print(classification_report(y_test,mnb_pred))
```
